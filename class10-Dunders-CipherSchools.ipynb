```python
class A:
  def __init__(self):
    print(self)
    print("initialized")

    def __del__(self):
      print(self)
      print("I am dying")
a = A()
```

```python
class Person:
  def __init__(self, name):
    self.name = name

  def say_hi(self):
    print('Hello, my name is', self.name)

p = Person('komal')
p.say_hi()
```
